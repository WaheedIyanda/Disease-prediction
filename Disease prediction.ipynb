# training the model

```python
Disease_dataa.columns
```

```python
d_dataa= Disease_dataa.drop('prognosis', axis=1)
```

```python
d_data=Disease_data.drop('prognosis', axis=1)
```

```python
X_train= d_dataa.iloc[:, :-1]
y_train= Disease_dataa['prognosis']
```

```python
X_test= d_data
y_test= Disease_data['prognosis']
```

```python
print(X_train.shape, X_test.shape)
```

```python
X_test.head(10)
```

```python
X_train.head(10)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
d_model= LogisticRegression()
```

```python
d_model.fit(X_train, y_train)
```

# checking out model training accuracy

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
d_prediction = d_model.predict(X_test)
```

```python
print (confusion_matrix (y_test, d_prediction))
print(classification_report(y_test, d_prediction))
```

```python
print(accuracy_score(y_test, d_prediction))
```
